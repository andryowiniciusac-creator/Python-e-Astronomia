```python
from astropy import units as u

distancia = 1 * u.au  # 1 Unidade Astronômica
print(distancia.to(u.km)) # O Astropy já sabe o valor exato da conversão
# O exemplo acima foi gerado 100% por IA para estudar a sintaxe
```
